```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request
from PIL import Image 
import os
```

```python
# open the url
html = urlopen('https://moonlighter.gamepedia.com/Enemies')
bs = BeautifulSoup(html, 'html.parser')
# find all img elements where .png in their 'src'
images = bs.find_all('img')

print(images[10])
```

```python
# print the url of each image scrapped
for image in images[1:-12]: 
    print(image['src']+'\n')

# check number of images
print(len(images))
```

```python
# function to download images based on their url
def download_png(url,file_path,file_name):
    full_path= file_path+file_name+'.png'
    # download the image from the url to the full_path
    urllib.request.urlretrieve(url,full_path)
```

```python
i=0
for img in images[1:-12]:
    try:
        download_png(img['src'],'moonlighter/',f'moonlighter_{i}')
        i+=1
    except:
        pass
```
