```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request
from PIL import Image
```

```python
# open the url
html = urlopen('https://castaway.fandom.com/wiki/List_of_Monsters_(Castaway_1_%26_2)?fbclid=IwAR3yi48aRwAR36CHofauA_MAMLi5203YRZFDEBT_cHfdzO9KPQ5WhJ0Ayw4')
bs = BeautifulSoup(html, 'html.parser')
# find all img elements where .png in their 'src'
images = bs.find_all('img', {'alt':re.compile('.png')})

# print the url of each image scrapped
for image in images: 
    print(image['src']+'\n')

# check number of images
print(f'I found {len(images)} monsters in Castaway Website')
```

```python
# Download the images based on their urls
```

```python
# function to download images based on their url
def download_png(url,file_path,file_name):
    full_path= file_path+file_name+'.png'
    # download the image from the url to the full_path
    urllib.request.urlretrieve(url,full_path)
```

```python
i=0
for img in images:
    try:
        download_png(img['src'],'castaway/',f'castaway_{i}')
        i+=1
    except:
        pass
```
