```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request
from PIL import Image 
import os
```

# Found urls of each monsters from website

```python
def download_png(url,file_path,file_name):
    full_path= file_path+file_name+'.png'
    # download the image from the url to the full_path
    urllib.request.urlretrieve(url,full_path)
```

```python
html = urlopen(f'https://dragonquest.fandom.com/wiki/List_of_monsters_in_Dragon_Quest_XI_Bestiary')
bs = BeautifulSoup(html, 'html.parser')
# find all img elements where .png in their 'src'
images = bs.find_all('img')
images=images[4:-3]

    
```

```python
for image in images: 
    print(image['data-src']+'\n')

# check number of images
print(len(images))
```

```python
i=508
for img in images:
    try:
        download_png(img['data-src'],'dragon_quest/',f'drag_{i}')
        i+=1
    except:
        print('failed')
        pass
```
