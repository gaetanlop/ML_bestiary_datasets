# Imports

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request
from PIL import Image 
```

# Found urls of each monsters from website

```python
# open the url
html = urlopen('https://dontstarve.fandom.com/wiki/Monster')
bs = BeautifulSoup(html, 'html.parser')
# find all img elements where .png in their 'src'
images = bs.find_all('img', {'data-src':re.compile('.png')})

# print the url of each image scrapped
for image in images: 
    print(image['data-src']+'\n')

# check number of images
print(f'I found {len(images)} monsters in Dontstarve Website')
```

# Download the images based on their urls

```python
# function to download images based on their url
def download_png(url,file_path,file_name):
    full_path= file_path+file_name+'.png'
    # download the image from the url to the full_path
    urllib.request.urlretrieve(url,full_path)
```

```python
i=0
for img in images:
    try:
        download_png(img['data-src'],'dontstarve/',f'dontstarve_{i}')
        i+=1
    except:
        pass
```
