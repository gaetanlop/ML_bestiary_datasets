# Imports

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import urllib.request
from PIL import Image 
import os
```

# Found urls of each monsters from website

```python
# open the url
html = urlopen('https://rosprites.blogspot.com/2012/05/monsters-all.html?fbclid=IwAR1_46qMZOhO1eBoQZHJs8YwU3Ni_618noSWxQpsF4kvNayoDahGgIplVO8')
bs = BeautifulSoup(html, 'html.parser')
# find all img elements where .png in their 'src'
images = bs.find_all('img', {'src':re.compile('.com')})

# print the url of each image scrapped
for image in images: 
    print(image['src']+'\n')

# check number of images
print(f'I found {len(images)} monsters gif in Ragnarok Website')
```

# Download the images based on their urls

```python
# function to download images based on their url
def download_png(url,file_path,file_name):
    full_path= file_path+file_name+'.gif'
    # download the image from the url to the full_path
    urllib.request.urlretrieve(url,full_path)
```

```python
i=0
for img in images:
    try:
        download_png(img['src'],'ragnarok/',f'ragnarok_{i}')
        i+=1
    except:
        pass
```

# Gif to png

```python
num=0
directory = 'ragnarok'

for filename in os.listdir(directory):
    im = Image.open(f'ragnarok/{filename}')
    im.save(f'ragnarok_png/ragna_png_{num}.png')
    num+=1
```
